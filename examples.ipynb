```python
import gpwrap
from IPython.display import Image
import numpy as np
```

```python
gp = gpwrap.GnuPlotWrapper()
```

```python
gp.clear()
dat1 = np.random.rand(200)
dat2 = np.random.rand(10)
gp.nokey()
gp.plot(dat1,"with points pt 7")
gp.plot(np.vstack([20*np.arange(len(dat2)),dat2]).T,"with lines lw 4")
Image(filename=gp.show())
```

```python
gp.clear()
gp.xlabel('x-label')
gp.ylabel('y-label')
gp.cm("plot sin(x), cos(0.4*x)")
Image(filename=gp.show())
```

```python
gp.clear()
script = """
set samples 51
set isosample 21
set ztics 0.5
splot [x=-3:3] [y=-3:3] sin(0.5*x) * cos(y)
"""
gp.cm(script)
Image(filename=gp.show())
```
